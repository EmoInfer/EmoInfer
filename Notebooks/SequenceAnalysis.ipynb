```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import sys
import subprocess
```

```python
df = pd.read_table('../extracted/extracted_classroom.csv', delimiter=',')
df.fillna('', inplace=True)
print(df)

emos = [            'Amusement',             'Happiness',                   'Awe',
                 'Pride',              'Surprise',                 'Anger',
              'Confused',              'Contempt',               'Disgust',
         'Embarrassment',                  'Fear',                  'Pain',
                 'Shame',              'Interest',               'Sadness',
     'Happily Surprised',     'Happily Disgusted',         'Sadly Fearful',
           'Sadly Angry',       'Sadly Surprised',       'Sadly Disgusted',
       'Fearfully Angry',   'Fearfully Surprised',   'Fearfully Disgusted',
     'Angrily Surprised',     'Angrily Disgusted', 'Disgustedly Surprised',
       'Appalled/Hatred']

for stri in reversed(emos):
#     df.replace({stri : emos.index(stri)}, regex=True)
    df = df.replace(regex=[stri], value = str(emos.index(stri) + 1))
    
# df.replace(regex=['Surprise'], value='he')
# df1.replace(regex=[' '], value='_')

print(df)
```

```python
f1 = open("../sequencing/Cordaro_raw_sequences.txt", 'w+')
f2 = open("../sequencing/Keltner_raw_sequences.txt", 'w+')
f3 = open("../sequencing/Du_raw_sequences.txt", 'w+')
```

```python
# COR, KEL,DU
fcs = np.array(df.face_id)
face_ids = np.unique(fcs)
# print(face_ids)

for face in face_ids:
    df1 = df.loc[df['face_id'] == face]
    flag1 = 0
    flag2 = 0
    flag3 = 0
    for ind, row in df1.iterrows():
#         print(row['Cordaro'])
        if row['Cordaro'] is not "":
            print('<' + str(row['frame']) + '> ' + ' '.join(row['Cordaro'].split(", ")) + ' -1 ')
            f1.write('<' + str(row['frame']) + '> ' + ' '.join(row['Cordaro'].split(", ")) + ' -1 ')
            flag1 = 1
            #         f1.write(' '.join(row['Cordaro'].split(", ")))
#         f1.write(' -1 ')
        if row['Keltner'] is not "":
            flag2 = 1
            f2.write('<' + str(row['frame']) + '> ' + ' '.join(row['Keltner'].split(", ")) + ' -1 ')
#         f2.write(' '.join(row['Keltner'].split(", ")))
#         f2.write(' -1 ')
        if row['Du'] is not "":
            flag3 = 1
            f3.write('<' + str(row['frame']) + '> ' + ' '.join(row['Du'].split(", ")) + ' -1 ')
#         f3.write(' '.join(row['Du'].split(", ")))
#         f3.write(' -1 ')
    if flag1:
        f1.write('-2\n')
    if flag2:
        f2.write('-2\n')    
    if flag3:
        f3.write('-2\n')    
        
f1.close()
f2.close()
f3.close()
```
